```python
#Import Libraries
import pandas as pd
import numpy as np
```

```python
#Question 1. Import the dataset into a pandas dataframe. Make sure that the date column is in pandas date time format.

# Load dataset
df = pd.read_csv("train.csv")

# Convert datetime column to pandas datetime format
df['datetime'] = pd.to_datetime(df['datetime'])

# Preview data
df.head()
```

```python
#Question 2. Check the data type of each column. How many rows are there in the dataset ? Does the dataset contain any missing values ?

# Check data types
print(df.dtypes)

# Number of rows
print("Number of rows:", len(df))

# Check missing values
print(df.isnull().sum())
```

```python
#Question 3. Using the date column, create new columns for: year, month, day of the week and hour of the day.

df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day_of_week'] = df['datetime'].dt.day_name()
df['hour'] = df['datetime'].dt.hour
```

```python
#Question 4. Rename the values in the season column to spring, summer, fall and winter.

season_mapping = {1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'}
df['season'] = df['season'].map(season_mapping)
```

```python
#Question 5. Calculate the total number of casual and registered bikes rented in the years 2011 and 2012.

total_by_year = df.groupby('year')[['casual', 'registered']].sum()
print(total_by_year)
```

```python
#Question 6. Calculate the mean of the hourly total rentals count by season. Which season has the highest mean ?

mean_by_season = df.groupby('season')['count'].mean()
print(mean_by_season)

# Find season with highest mean
highest_season = mean_by_season.idxmax()
print("Season with highest mean rentals:", highest_season)
```

```python
#Question 7. Are more bikes rented by registered users on working or non-working days ?  Does the answer differ for non-registered users ? Is the answer the same for both years ?

# For registered users
reg_by_workday = df.groupby('workingday')['registered'].mean()
print("Registered users rentals by working day:\n", reg_by_workday)

# For casual users
casual_by_workday = df.groupby('workingday')['casual'].mean()
print("Casual users rentals by working day:\n", casual_by_workday)

# Check if it differs by year
reg_by_year_workday = df.groupby(['year','workingday'])['registered'].mean()
casual_by_year_workday = df.groupby(['year','workingday'])['casual'].mean()
```

```python
#Question 8. Which months in the year 2011 have the highest and the lowest total number of bikes rented ? Repeat for the year 2012.

for yr in [2011, 2012]:
    month_totals = df[df['year']==yr].groupby('month')['count'].sum()
    print(f"Year {yr} - Month with highest rentals: {month_totals.idxmax()}, Lowest: {month_totals.idxmin()}")
```

```python
#Question 9. Which type of weather have the highest and lowest mean of the hourly total rentals count ?

mean_by_weather = df.groupby('weather')['count'].mean()
print(mean_by_weather)
print("Weather with highest mean rentals:", mean_by_weather.idxmax())
print("Weather with lowest mean rentals:", mean_by_weather.idxmin())
```

```python
#Question 10. Calculate the correlation between the hourly total rentals count and all the numerical columns in the dataset. Which column has the highest correlation with the total rentals count ?

numeric_cols = df.select_dtypes(include=np.number)  # selects only numeric columns
correlation = numeric_cols.corr()['count'].sort_values(ascending=False)
print(correlation)
print("Highest correlation with total rentals:", correlation.index[1], "=", correlation[1])
```

```python
#Question 11. Create a new categorical column called day_period, which can take four possible values: night, morning, afternoon and evening. These values correspond to the following binning of the hour column: 0-6: night, 6-12: morning, 12-6: afternoon, 6-24:evening.

def get_day_period(hour):
    if 0 <= hour < 6:
        return 'night'
    elif 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    else:
        return 'evening'

df['day_period'] = df['hour'].apply(get_day_period)
```

```python
#Question 12. Generate a pivot table for the mean of the hourly total rentals count, with the index set to the day period and the column set to the working day column. What can you observe from the table ?

pivot_table = df.pivot_table(values='count', index='day_period', columns='workingday', aggfunc='mean')
print(pivot_table)
```
